#__Ejericio 1: Manipulación de archivos__

Manipularemos los famosos Dataframes

```python
#Importar datos de el dataset de heart desease https://archive.ics.uci.edu/dataset/45/heart+disease

# fetch dataset
heart_disease = fetch_ucirepo(id=45)

# data (as pandas dataframes)
X = heart_disease.data.features
y = heart_disease.data.targets

# metadata
#print(heart_disease.metadata)

# variable information
#print(heart_disease.variables)
```

```python
df = pd.concat([X, y], axis=1)
```

* Explorar el dataset cargado

```python
df.head()
```

```python
df.info()
```

```python
def contar_nulos(df, columna):
    contador = 0
    for i in df[columna]:
        if math.isnan(i):
            contador+=1
            #print(i)
        else:
            continue#contador+=1
    return contador
```

```python
contar_nulos(df, 'ca')
```
