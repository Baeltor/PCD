## __Operaciones básicas__

`Matplotlib` es la API de `pyplot` predeterminada y tiene una interfaz global con un estilo muy similar a MATLAB

```python
x = np.arange(-10, 11)
x
```

```python
y = np.arange(-10+15, 11+15)
y
```

```python
plt.figure(figsize=(8, 4))

plt.title('My Nice Plot $\int$')

y = x ** 2
plt.plot(x, y, color='#b7950b', label='$x^2$', marker='>')

y = -1 * (x ** 2)
plt.plot(x, y, label='$-x^2$', color='r', marker='x')

plt.xlabel('$x^2$ - axis')
plt.ylabel('$y$- axis')
plt.legend()
plt.grid()
#plt.axis([-5, 5, -50, 50])
plt.show()
```

Podemos mejorar el estilo de los títulos agregando los parámetros de la fuente como un diccionario

```python
font = {'family':'fantasy','color':'#cb4335','size':20}

plt.figure(figsize=(8, 4))
plt.title('My Nice Plot', fontdict=font)
plt.plot(x, x ** 2, label= "$x^2$")
plt.plot(x, -1 * (x ** 2), label='$-x^2$', color='r')
plt.xlabel('x', fontdict=font)
plt.ylabel('$y = x^2$', fontdict=font)
plt.legend()
plt.grid()
plt.show()
```

Es posible generar varias gráficas de forma simultanea utilizando el comando `plt.subplot(rows,columns, panel_selected)`

```python
plt.figure(figsize=(16, 9))
plt.title('My Nice Plot')

plt.subplot(1, 2, 1)  # rows, columns, panel selected
plt.plot(x, x ** 2)
plt.plot([0, 0, 0], [-10, 0, 100])
plt.legend(['$x^2$', 'Vertical Line'])
plt.xlabel('x')
plt.ylabel('$y = x^2$')
plt.grid()
plt.title('A nice plot')

plt.subplot(1, 2, 2)
plt.plot(x, -1 * (x ** 2))
plt.plot([-10, 0, 10], [-50, -50, -50], color='green')
plt.legend(['$-x^2$', 'Horizontal Line'])
plt.title('Another nice plot')
plt.xlabel('$x$')
plt.ylabel('$Y = x^2$')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(16, 9))
plt.title('My Nice Plot')

for i in range(6):
    plt.subplot(2, 3, i + 1)
    #plt.subplot(1, 2, 1)  # rows, columns, panel selected
    y = x ** (i+1)
    plt.plot(x, y)
    plt.plot([0, 0, 0], [min(y), 0, max(y)])
    plt.legend([f'$x^{i+1}$', 'Vertical Line'])
    plt.xlabel('x')
    plt.ylabel(f'$y = x^{i+1}$')
    plt.grid()
    plt.title('A nice plot')
```

Podemos acceder a los objetos `fig` y `axes` a través del método `plt.subplots()`. Incluso podemos generar de manera automática un conjunto de gráficas utilizando el objeto axes y un ciclo simple

```python
import random

rows = 3
columns = 3
fig, axes = plt.subplots(rows, columns, figsize=(17, 10))
fig.suptitle('**Nice linear functions**', size=28, weight=25, color='r')

for row in range(rows):
  for column in range(columns):
    slope = random.randint(-5, 5)
    intercept = round(random.random(),1)
    my_figure = axes[row][column]
    my_figure.plot(x, slope * x + intercept , label=f'$y = {slope}x + {intercept}$', color=np.random.rand(3))
    my_figure.grid()
    my_figure.legend()
    my_figure.set_xlabel('$x$')
    my_figure.set_ylabel('$y$')
    my_figure.set_title(f'Figure ({row},{column})')

```

Podemos utilizar diferentes tipos de línea en nuestros gráficos. Podemos consultar un lista completa de los tipos de línea disponible en: [https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html](https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html)

```python
fig, axes = plt.subplots(figsize=(12, 6))

axes.plot(x, x + 0, linestyle='solid')
axes.plot(x, x + 1, linestyle='dashed')
axes.plot(x, x + 2, linestyle='dashdot')
axes.plot(x, x + 3, linestyle='dotted');

axes.set_title("My Nice Plot")
plt.grid()
plt.show()
```

Tambien diferentes marcadores para los puntos. Podemos consultar la lista completa de los marcadores disponibles en :[https://matplotlib.org/stable/api/markers_api.html](https://matplotlib.org/stable/api/markers_api.html)

```python
fig, axes = plt.subplots(figsize=(12, 6))

axes.plot(x, x + 0, marker='<')
axes.plot(x, x + 1, marker='X')
axes.plot(x, x + 2, marker='^')
axes.plot(x, x + 3, marker='P');

axes.set_title("My Nice Plot")
plt.grid()
plt.show()
```

Podemos utilizar cotracciones para combinar estilos

```python
fig, axes = plt.subplots(figsize=(12, 6))

axes.plot(x, x + 0, '-og', label="solid green")
axes.plot(x, x + 1, '--c', label="dashed cyan")
axes.plot(x, x + 2, '-Xy', label="dashdot blue")
axes.plot(x, x + 3, ':r', label="dotted red")

axes.set_title("My Nice Plot")

axes.legend()
plt.grid()
```

### __Dispersión__

```python
N = 50
#error = np.random.normal()
x1 = np.random.rand(N)
y1 = np.random.rand(N)
```

El gráfico de dispersión es uno de los más comunes y representa con un punto cada observación

```python
plt.figure(figsize=(14, 8))
plt.scatter(x1, y1, color='b', marker='o', s=80)
plt.grid()
```

Podemos combinar varios conjuntos de datos en un solo gráfico

```python
N = 50
x2 = np.random.rand(N)
y2 = np.random.rand(N)

plt.figure(figsize=(14, 8))
plt.scatter(x1, y1, color='r', marker='o', s=280, alpha=0.1)
plt.scatter(x2, y2, color='g', marker='<', s=80, alpha=0.7, edgecolors='k')
plt.plot([min(x1), max(x2)], [min(y1), max(y2)], color='b', linestyle='--')
plt.grid()
plt.title('Scatter Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['Group 1', 'Group 2'])
```

`Matplotlib` tiene una gran cantidad de mapas de colores.  Un mapa de colores es una especie de lista de colores en la cual cada color está ubicado en un rango entre `1` y `100`. Este es un ejemplo:

imagen.png

Podemos usar estos mapas para configurar un color variable en nuestros gráficos. Adicionalmente, si queremos agregar la barra de colores podemos utilizar el comando `plt.colorbar()`

```python
colors = np.random.randint(0, 100, size=N)

plt.figure(figsize=(14, 8))
plt.scatter(x1, y1, c=colors, marker='o', s=100, cmap='viridis')
plt.title('A fancy scatter plot using color maps')
plt.colorbar()
plt.grid()
```

También es posible realizar configuraciones más elaboradas especificando valores variables para el área de cada punto con el atributo `s` y así como el nivel de trasparencia con el atributo `alpha`

```python
plt.figure(figsize=(14, 8))

area = np.pi * (20 * np.random.rand(N))**2  # 0 to 15 point radii

plt.scatter(x1, y1, s=area, c=colors, alpha=0.5, cmap='Dark2')
plt.colorbar()

plt.show()
```

De igual forma, podemos generar múltiples gráficos con diferentes parámetros de configuración

```python
fig = plt.figure(figsize=(20, 8))

ax1 = fig.add_subplot(1,2,1)
plt.scatter(x1, y1, s=area, c=colors, alpha=0.2, cmap='rainbow')
plt.colorbar()
plt.show()

ax2 = fig.add_subplot(1,2,2)
plt.scatter(x2, y2, s=area, c=colors, alpha=0.8, cmap='gnuplot')
plt.colorbar()

plt.show()
```

### __Histogramas__

Otro gráfico de utilidad son los histogramas. Estos nos permiten comprobar la distribución de frecuencias de un conjunto de datos. El siguiente ejemplo muestra un histograma construido a partir de una muestra aleatoria de una distribución normal.  El parámetro `bins` nos permite ajustar el valor de los intervalos para el hsitograma

```python
values = np.random.randn(10000)
```

```python
help(plt.hist)
```

```python
plt.subplots(figsize=(14, 8))

plt.hist(values, bins='auto', alpha=0.8, histtype='bar', color='steelblue', edgecolor='green')
plt.xlim(xmin=-5, xmax=5)

plt.show()
```

Podemos utilizar scipy para estimar la función de densidad de la muestra aleatoria mediate el método `stats.kde.gaussian_kde()`

```python
from scipy import stats

density = stats.kde.gaussian_kde(values)
```

Podemos visualizar la función de desinsidad de probabilidad estimada de la siguiente forma

```python
plt.subplots(figsize=(12, 6))

values2 = np.linspace(min(values) - 5, max(values) + 5, 100)

plt.plot(values2, density(values2), color='orange')
plt.fill_between(values2, 0, density(values2), alpha=0.1, color='orange')
plt.xlim(xmin=-5, xmax=5)

plt.show()
```

Finalmente, podemos comparar el histograma con la función de densidad estimada.

```python
plt.subplots(figsize=(14, 8))

plt.hist(values, bins='auto', alpha=0.8, density=True, histtype='bar', color='steelblue', edgecolor='green')
plt.plot(values2, density(values2), color='orange', linewidth=3.0)

plt.xlim(xmin=-5, xmax=5)
plt.grid()
plt.show()
```

### __Diagrama de cajas__

Los diagramas de cajas nos permiten representar un conjunto de datos a partir de sus cuartiles. Son especialmente útiles para visualizar posibles datos atípicos.


imagen.png

```python
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(80, 30, 200)
data3 = np.random.exponential(100, 200)
data4 = np.random.normal(70, 25, 200)

data_to_plot = [data1, data2, data3, data4]
```

```python
plt.figure(figsize=(15, 9))
plt.boxplot(data_to_plot)
plt.show()
```

Como en todos los casos anteriores tenemos diferentes formas para ajustar los parámetros de los gráficos

```python
plt.figure(figsize=(15, 9))
flierprops = dict(marker='o', markerfacecolor='green', markersize=10, linestyle='none', )
plt.boxplot(data_to_plot, flierprops=flierprops)
plt.show()
```

```python
plt.figure(figsize=(15, 9))
flierprops = dict(marker='o', markerfacecolor='green', markersize=10, linestyle='none')
plt.boxplot(data_to_plot, flierprops=flierprops, vert=False, patch_artist=True )
plt.show()
```

*   `hexbin`
*   `hist`
*   `kde`
*   `line`
*   `pie`

El valor por defecto para el atributo `kind` es el tipo de gráfico `line`. Adicionalmente podemos utilizar muchos de los atributos de configuración que usamos en `matplotlib`

```python
df.plot(x="Rank", y="P75th", figsize=(14,8), grid=True, color='r', marker='o', linestyle="dashed", title="A panda line plot",)
plt.show()
```
