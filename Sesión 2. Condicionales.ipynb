Los condicionales nos permiten controlar el flujo de ejecución de un programa a partir del cumplimiento de una condición. Al evaluar dicha condición el resultado debe ser siempre `Verdadero` o `Falso`.

### __Indentación__

En Python, la indentación es de caracter obligatorio y se refiere al número de espacios al principio de una línea de código. Esta nos permite separar los bloques de código y mantener una mejor estructura y legibilidad  de nuestros programas (usualmente se usan 2 espacios para indicar un nuevo bloque de instrucciones)

```python
5 > 2
```

```python
if 5 > 2:
    print("5 es mayor que 2")
```

Si no realizamos la indentación correctamente python generará un error

```python
if 5 > 2:
print("5 es mayor que 2")
```

Se debe garantizar la misma cantidad de espacios en un bloque de código, de otra forma Python generará un error.

```python
if 5 > 2:
    print("5 es mayor que 2")

    if 5 > 2:
        print("5 es mayor que 2")
```

### __Tipos de condicionales__

Python nos permite crear condicionales de 3 tipos:

*   `<If Statement>`
*   `<If Else Statement>`
*   `<Nested If>`

__Ejemplo `<If Statement>`__: determinar si hay sobrecosto por equipaje demasiado pesado

```python
peso = float(input('Cuantos Kilos lleva de equipaje: '))
if peso > 50:
  print('Se debe pagar un sobrecosto del 10% por llevar equipaje pesado')
print('Continue a la taquilla')
```

__Ejemplo `<If Else Statement>`__: Encontrar el valor absoluto de un número:

```python
a = int(input('Ingrese un número: '))

if a >= 0:
  print(f'|{a}| = {a}')
else:
  print(f'|{a}| = {-a}')
```

Si la expresión es corta, es posible escribir condicionales de tipo  `<If Else Statement>` utilizando una contracción

```python
a = int(input("Por favor ingrese un número"))

print(f'|{a}| = {a}') if a > 0 else print(f'|{a}| = {-a}')
```

__Ejemplo `<Nested If>`__: encontrar el mayor de 2 números

```python
a = -12
b = 3

if (a < b and a + b > 0):
  print(f"El mayor es {b}")
elif (a+b<=0):
  print(f'El mayor es {b} y además {a} es negativo')
elif a > b:
   print(f"El mayor es {a}")
else:
  print(f"Los números son iguales")
```

Utilizando una contracción:

```python
a = 1
b = -1
print(f'El mayor es: {a}' if  a > b else f'El mayor es: {b}' if  a < b else 'Son iguales')
```

### **Ejercicio 2: calcular el IMC**

Asuma que el riesgo de que una persona sufra problemas del corazón depende de su edad y de su índice de masa corporal (IMC)

|          | Edad $<$ 45     | Edad $\geq 45$ |
|--------------|-----------|------------|
| IMC $<$ 22    | Bajo      | Medio        |
| IMC $\geq 22$ | Medio  | Alto       |

El índice de masa corporal (IMC) es el cociente entre el peso del individuo en kilos y el cuadrado de su altura en metros. Escriba un programa que reciba como entrada la altura, el peso y la edad de una persona, y entregue su IMC y su nivel actual del riesgo.

```python
# Solución propuesta
edad = int(input("Por favor ingrese su edad en años: "))
peso = float(input("Por favor ingrese su peso en kg: "))
altura = float(input("Por favor ingrese su altura en metros: "))
IMC = peso/(altura**2)
print(edad, peso, altura, IMC)

if edad < 45:
    if IMC < 22:
        print('El riesgo de tener problemas de corazon es bajo')
    else:
        print('El riesgo de tener problemas de corazon es medio')
else:
    if IMC < 22:
        print('El riesgo de tener problemas de corazon es medio')
    else:
        print('El riesgo de tener problemas de corazon es alto')
```
